```python
#경우의 수, 찾으신금액이 10만 미만이면 소비, 10만~20만 사이이면 과소비, 20만 이상
#이면 파산
def cal(x):
  if 100000>=x:
    return '소비'
  elif 200000>=x:     #x>100000 안넣어도 됨. elif라서.
    return '과소비'
  else:
    return '파산'
df2['찾으신금액'].map(cal)

#선생님 풀이
df2['소비기준']=''
df2.loc[df2['찾으신금액']<100000,'소비기준']='적음'
#이렇게 해도 되지만,

#내가 푼 방법으로 하심.
```

```python
#apply 활용

#컬럼에 대한 일괄 연산
def get_col_data(col):
  return col.sum()

df2[['찾으신금액','맡기신금액']].apply(get_col_data,axis=0)
def get_row_data(row):
  return row.mean()
df2[['찾으신금액','맡기신금액']].apply(get_row_data,axis=1)
def introduce(row):
  return f"{row['거래일시']}에 {row['기재내용']}을 하기위해 {row['찾으신금액']}을 사용해서 {row['거래후 잔액']}이 남았습니다. {row['소비기준']} 소비였습니다."

df2.apply(introduce,axis=1)
```
