```python
df['번호']=df['번호'].astype('str')+'번'
```

```python
def string_change(s):
  return s.zfill(4)
#df['번호'].str.zfill(4)
df['번호'].apply(string_change)
```

```python
# 2.생년월일에 년, 월, 일 붙이기
# 2050-01-01 => 2050년01월01일
df['생년월일']=df['생년월일'].astype('str')
def put_word(s):
  return s[:4]+'년'+s[5:7]+'월'+s[8:10]+'일'
df['생년월일'].apply(put_word)
```

```python
# 3. apply를 이용해 평균과 총점 구하기
def get_avg(col):
  return col.mean()
def get_sum(col):
  return col.sum()
df2=df[['수학','영어','국어']]
df2.apply(get_avg)
df2.apply(get_sum)
```

```python
# 4. 학업성취도를 아래와 같이 나눠서 새로운 컬럼으로 만들기.(평균 기준)
## 90 이상이면 최우수
## 80 ~ 90 사이면 우수
## 70 ~ 80 사이면 보통
## 60 ~ 70 사이면 미달

df['평균']=(df['국어']+df['영어']+df['수학'])/3
df['등급']='보통'
def grading(num):
  if num>=90:
    return '최우수'
  elif num>=80 and num<90:
    return '우수'
  elif num>=70 and num<80:
    return '보통'
  else:
    return '미달'

df['등급']=df['평균'].apply(grading)
df
```

```python
# 5. apply를 이용해 min-max scaler 일괄 적용
# 어떤 데이터군을 0 ~ 1 사이의 값으로 범위를 한정시키는 방법이다. 크기 기준이 제각각 다른 데이터간 기준을 맞추어 관계를 명확히 표현하기 위해서 사용한다.
# min - max scale 공식은 아래와 같다.
# 어떤 데이터군(리스트, 시리즈)이 있을 때, 어떤 대상값을 0 ~ 1 사이의 값으로 만들려면
# (대상값 - 해당 데이터 군의 최소값) / (해당 데이터 군의 최대값 - 해당 데이터 군의 최소값)
def min_max(col):

  return (col-col.min())/(col.max()-col.min())
min_max(df['영어'])
```

```python
# 6. applymap을 이용해 모든 점수 뒤에 점 붙이기
# 90점, 91점 ...
def put_letter(x):

  return str(x)+'점'
df['국어'].apply(put_letter)
df[['국어','수학','영어']].applymap(put_letter)
```
